```python
#3.2
#Problem:1
import pandas as pd

# Load dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# Fill missing Age with median
median_age = titanic['Age'].median()
titanic['Age'].fillna(median_age, inplace=True)

# Add fare_per_year column
titanic['fare_per_year'] = titanic['Fare'] / titanic['Age']

# Filter rows where fare_per_year > 5
high_fare_age = titanic[titanic['fare_per_year'] > 5]

# Sort descending by fare_per_year
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# Select only Name and fare_per_year columns
result = high_fare_age_srt[['Name', 'fare_per_year']]

# Show the result
print(result)
```

```python
#3.2
#Problem:2
import pandas as pd

# Load dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# Fill missing Age values with median
median_age = titanic['Age'].median()
titanic['Age'].fillna(median_age, inplace=True)

# Step 1: Add fare_per_class column
titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']

# Step 2: Subset adult males
adult_males = titanic[(titanic['Age'] >= 18) & (titanic['Sex'] == 'male')]

# Step 3: Sort descending by fare_per_class
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

# Step 4: Select Name, Age, fare_per_class columns
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]

# Step 5: View the result
print(result)

```

```python
#3.3
#Problem:1
import pandas as pd

# Load dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# Step 1: Total fare revenue
total_fare = titanic['Fare'].sum()
print("Total Fare Revenue:", total_fare)

# Step 2: Total fare per class
fare_class1 = titanic[titanic['Pclass'] == 1]['Fare'].sum()
fare_class2 = titanic[titanic['Pclass'] == 2]['Fare'].sum()
fare_class3 = titanic[titanic['Pclass'] == 3]['Fare'].sum()

# Step 3: Combine totals into a list
fare_totals = [fare_class1, fare_class2, fare_class3]
print("Fare Totals by Class:", fare_totals)

# Step 4: Calculate percentage of total fare revenue by class
fare_by_class = titanic.groupby('Pclass')['Fare'].sum()
fare_percent = (fare_by_class / fare_by_class.sum()) * 100
print(fare_percent)
```

```python
#3.3
#Problem:2
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
titanic = pd.read_csv("Titanic-Dataset.csv")

# Fill missing Age with median
median_age = titanic['Age'].median()
titanic['Age'].fillna(median_age, inplace=True)

# Step 1: Create age_group column
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

titanic['age_group'] = titanic['Age'].apply(categorize_age)

# Step 2: Total passengers
total_passengers = titanic.shape[0]

# Step 3: Count passengers in each age group
age_group_counts = titanic['age_group'].value_counts()

# Step 4: Percentage of passengers
age_group_percent = (age_group_counts / total_passengers) * 100

# Display results
print("Percentage of Passengers by Age Group:")
print(age_group_percent)

# Step 5: Pie chart
plt.figure(figsize=(6,6))
age_group_percent.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Percentage of Passengers by Age Group')
plt.ylabel('')  # Hide y-label
plt.show()
```
