```python
#TO-Do-3:

# Problem 1: Define the train-test split function from scratch
def train_test_split_scratch(x, y, test_size=0.3, random_seed=42):
    """
    Splits dataset into training and testing sets without using sklearn.

    Parameters:
    x : np.array - Features
    y : np.array - Labels
    test_size : float - Proportion of test set (default 0.3)
    random_seed : int - Seed for reproducibility (default 42)

    Returns:
    X_train, X_test, Y_train, Y_test
    """
    np.random.seed(random_seed)  # For reproducibility

    #  Shuffle indices
    indices = np.arange(x.shape[0])
    np.random.shuffle(indices)

    # Calculate test set size
    test_split_size = int(len(x) * test_size)

    # Split indices
    test_indices = indices[:test_split_size]
    train_indices = indices[test_split_size:]

    # Split data into training and testing sets
    X_train, X_test = x[train_indices], x[test_indices]
    Y_train, Y_test = y[train_indices], y[test_indices]

    return X_train, X_test, Y_train, Y_test

# Problem 2: Use the function (80–20 split)
X_train, X_test, Y_train, Y_test = train_test_split_scratch(X, Y, test_size=0.2)


print("\nAfter train-test split:")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)

```

```python
#TO-DO_4

import numpy as np

# Define the cost function
def cost_function(X, Y, W):
    """
    Parameters:
    This function finds the Mean Squared Error.

    Input parameters:
    X: Feature Matrix (n_samples × n_features)
    Y: Target Vector (n_samples,)
    W: Weight Vector (n_features,)

    Output Parameters:
    cost: accumulated mean squared error
    """
    # Convert inputs to numpy arrays and reshape if needed
    X = np.array(X, dtype=float)
    Y = np.array(Y, dtype=float).reshape(-1, 1)  # Column vector
    W = np.array(W, dtype=float).reshape(-1, 1)  # Column vector

    n = len(Y)  # Number of samples

    # Step 1: Compute predictions
    Y_predicted = X @ W  # Matrix multiplication

    # Step 2: Compute error
    error = Y_predicted - Y

    # Step 3: Compute cost (Mean Squared Error)
    cost = (1 / (2 * n)) * np.sum(error ** 2)

    return cost


    # Example data and function call (put this AFTER the function)
X = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [3, 4, 5],
    [4, 5, 6],
    [5, 6, 7]
])
Y = np.array([10, 12, 14, 16, 18])
W = np.array([0.2, 0.4, 0.6])

mse = cost_function(X, Y, W)
print("Mean Squared Error:", mse)
```

```python
import numpy as np

# ---------------- TO-DO 8: RMSE Function ----------------
def rmse(Y, Y_pred):
    """
    This function calculates the Root Mean Squared Error (RMSE).

    Parameters:
    Y: Array of actual target values (n_samples,)
    Y_pred: Array of predicted target values (n_samples,)

    Returns:
    rmse_value: Root Mean Squared Error
    """
    Y = np.array(Y, dtype=float)
    Y_pred = np.array(Y_pred, dtype=float)

    # Step 1: Compute squared differences
    error = Y - Y_pred
    squared_error = error ** 2

    # Step 2: Compute mean squared error
    mse = np.mean(squared_error)

    # Step 3: Take square root
    rmse_value = np.sqrt(mse)

    return rmse_value

# ---------------- Example/Test ----------------
# Actual values
Y_actual = np.array([3, 5, 7, 9, 11])
# Predicted values (for example, using some model)
Y_predicted = np.array([2.9, 5.1, 7.0, 8.8, 11.2])

# Compute RMSE
rmse_value = rmse(Y_actual, Y_predicted)
print("RMSE:", rmse_value)
```

```python
import numpy as np

# ---------------- TO-DO 9: R-Squared Function ----------------
def r2(Y, Y_pred):
    """
    This function calculates the R Squared (coefficient of determination).

    Parameters:
    Y: Array of actual target values (n_samples,)
    Y_pred: Array of predicted target values (n_samples,)

    Returns:
    r2_value: R Squared Error
    """
    Y = np.array(Y, dtype=float)
    Y_pred = np.array(Y_pred, dtype=float)

    # Step 1: Mean of actual values
    mean_y = np.mean(Y)

    # Step 2: Total sum of squares
    ss_tot = np.sum((Y - mean_y) ** 2)

    # Step 3: Residual sum of squares
    ss_res = np.sum((Y - Y_pred) ** 2)

    # Step 4: R-Squared
    r2_value = 1 - (ss_res / ss_tot)

    return r2_value

# ---------------- Example/Test ----------------
# Actual values
Y_actual = np.array([3, 5, 7, 9, 11])
# Predicted values (from a model)
Y_predicted = np.array([2.9, 5.1, 7.0, 8.8, 11.2])

# Compute R-Squared
r2_value = r2(Y_actual, Y_predicted)
print("R-Squared:", r2_value)
```

```python
#TO-DO 10:
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# ---------------- TO-DO 4: Cost Function ----------------
def cost_function(X, Y, W):
    X = np.array(X, dtype=float)
    Y = np.array(Y, dtype=float).reshape(-1, 1)
    W = np.array(W, dtype=float).reshape(-1, 1)
    n = len(Y)
    Y_pred = X @ W
    error = Y_pred - Y
    cost = (1 / (2 * n)) * np.sum(error ** 2)
    return cost

# ---------------- TO-DO 6: Gradient Descent ----------------
def gradient_descent(X, Y, W, alpha, iterations):
    X = np.array(X, dtype=float)
    Y = np.array(Y, dtype=float).reshape(-1, 1)
    W_update = np.array(W, dtype=float).reshape(-1, 1)
    m = len(Y)
    cost_history = []

    for iteration in range(iterations):
        Y_pred = X @ W_update
        loss = Y_pred - Y
        dw = (1 / m) * (X.T @ loss)
        W_update = W_update - alpha * dw
        cost_history.append(cost_function(X, Y, W_update))

    return W_update.flatten(), cost_history  # Flatten W for easier usage

# ---------------- TO-DO 8: RMSE ----------------
def rmse(Y, Y_pred):
    Y = np.array(Y, dtype=float)
    Y_pred = np.array(Y_pred, dtype=float)
    mse = np.mean((Y - Y_pred) ** 2)
    return np.sqrt(mse)

# ---------------- TO-DO 9: R-Squared ----------------
def r2(Y, Y_pred):
    Y = np.array(Y, dtype=float)
    Y_pred = np.array(Y_pred, dtype=float)
    ss_tot = np.sum((Y - np.mean(Y)) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)

# ---------------- Main Function ----------------
def main():
    # Step 1: Load dataset
    df = pd.read_csv("/content/drive/MyDrive/Ai/student.csv")

    # Step 2: Prepare features and target
    X = df[['Math', 'Reading']].values  # Features
    Y = df['Writing'].values            # Target

    # Step 3: Split into training and test sets (80%-20%)
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    # Step 4: Initialize weights, learning rate, iterations
    W_init = np.zeros(X_train.shape[1])
    alpha = 0.00001
    iterations = 1000

    # Step 5: Perform Gradient Descent
    W_optimal, cost_history = gradient_descent(X_train, Y_train, W_init, alpha, iterations)

    # Step 6: Predict on test set
    Y_pred = X_test @ W_optimal

    # Step 7: Evaluate model
    model_rmse = rmse(Y_test, Y_pred)
    model_r2 = r2(Y_test, Y_pred)

    # Step 8: Output results
    print("Final Weights:", W_optimal)
    print("Cost History (First 10 iterations):", cost_history[:10])
    print("RMSE on Test Set:", model_rmse)
    print("R-Squared on Test Set:", model_r2)

# Execute the main function
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------- RMSE function ----------------
def rmse(Y_true, Y_pred):
    return np.sqrt(np.mean((Y_true - Y_pred) ** 2))

# ---------------- Gradient Descent function ----------------
def gradient_descent(X, Y, W, alpha=0.01, iterations=1000):
    m = X.shape[0]
    cost_history = []

    W_update = W.copy()

    for i in range(iterations):
        Y_pred = X @ W_update
        loss = Y_pred - Y
        dw = (1 / m) * (X.T @ loss)
        W_update = W_update - alpha * dw
        cost_history.append(np.mean(loss ** 2))

    return W_update, cost_history

# ---------------- TO-DO 11 Function ----------------
def to_do_11(X_train, Y_train, X_test, Y_test, W_init, iterations=1000):
    # Ensure proper shape
    Y_train = Y_train.reshape(-1,1) if len(Y_train.shape)==1 else Y_train
    Y_test = Y_test.reshape(-1,1) if len(Y_test.shape)==1 else Y_test
    W_init = W_init.reshape(-1,1) if len(W_init.shape)==1 else W_init

    # 1) Model Performance Check
    alpha = 0.00001
    W_opt, cost_history = gradient_descent(X_train, Y_train, W_init, alpha, iterations)
    Y_train_pred = X_train @ W_opt
    Y_test_pred = X_test @ W_opt

    train_rmse = rmse(Y_train, Y_train_pred)
    test_rmse = rmse(Y_test, Y_test_pred)

    print("1. Model Performance Check:")
    print("Train RMSE:", round(train_rmse,4))
    print("Test RMSE:", round(test_rmse,4))

    if abs(train_rmse - test_rmse) < 0.5:
        print("Performance is acceptable: No overfitting or underfitting.\n")
    elif train_rmse < test_rmse:
        print("Potential Overfitting.\n")
    else:
        print("Potential Underfitting.\n")

    # 2️)Learning Rate Experiment
    print("2. Learning Rate Experiment:")
    learning_rates = [0.000001, 0.00001, 0.0001, 0.001]
    plt.figure(figsize=(10,6))

    for lr in learning_rates:
        W_temp, cost_hist_temp = gradient_descent(X_train, Y_train, W_init, alpha=lr, iterations=iterations)
        Y_test_pred_temp = X_test @ W_temp
        print(f"Alpha = {lr:.6f}, Test RMSE = {rmse(Y_test, Y_test_pred_temp):.4f}, Final Cost = {cost_hist_temp[-1]:.6f}")
        plt.plot(cost_hist_temp, label=f'alpha={lr:.6f}')

    plt.title("Cost vs Iterations for Different Learning Rates")
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    plt.legend()
    plt.show()


# ---------------- Example Usage ----------------
# Generate example data
np.random.seed(0)  # for reproducibility

# 100 training samples, 5 features
X_train = np.random.rand(100, 5)
true_weights = np.array([[1.5], [-2.0], [0.5], [1.0], [-1.0]])
Y_train = X_train @ true_weights + np.random.randn(100,1)*0.1  # add some noise

# 20 test samples, 5 features
X_test = np.random.rand(20, 5)
Y_test = X_test @ true_weights + np.random.randn(20,1)*0.1  # add some noise

# Initialize weights
W_init = np.zeros((X_train.shape[1], 1))

# Call your existing function
to_do_11(X_train, Y_train, X_test, Y_test, W_init, iterations=1000)
```
