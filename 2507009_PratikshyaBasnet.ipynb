```python
#Task 3 => Outlier Detection
south_asia_df['Outlier'] = False

for year in south_asia_df['year'].unique():
    temp = south_asia_df[south_asia_df['year']==year]

    #computing IQR for HDI
    Q1_hdi = temp['hdi'].quantile(0.25)
    Q3_hdi = temp['hdi'].quantile(0.75)
    IQR_hdi = Q3_hdi - Q1_hdi

    #computing IQR for GNI
    Q1_gni = temp['gross_inc_percap'].quantile(0.25)
    Q3_gni = temp['gross_inc_percap'].quantile(0.75)
    IQR_gni = Q3_gni - Q1_gni

    #Appling 1.5* IQR rule in HDI
    hdi_low = Q1_hdi - 1.5*IQR_hdi
    hdi_high = Q3_hdi + 1.5*IQR_hdi

    #Appling 1.5* IQR rule in GNI
    gni_low = Q1_gni - 1.5*IQR_gni
    gni_high = Q3_gni + 1.5*IQR_gni

    #Detecting Outlier in HDI and GNI
    mask = ((temp['hdi'] < hdi_low) | (temp['hdi'] > hdi_high) |
            (temp['gross_inc_percap'] < gni_low) | (temp['gross_inc_percap'] > gni_high))
    south_asia_df.loc[temp.index, 'Outlier'] = mask

#Plotting Outliers for 2022
df_year = south_asia_df[south_asia_df['year']==2022]
outliers = df_year[df_year['Outlier']]
normal = df_year[~df_year['Outlier']]

plt.figure(figsize=(10,6))
plt.scatter(normal['gross_inc_percap'], normal['hdi'],
            color='blue', s=80, alpha=0.6, label='Normal Countries')
plt.scatter(outliers['gross_inc_percap'], outliers['hdi'],
            color='red', s=150, alpha=0.8, label='Outliers',
            edgecolors='black', linewidths=1.5)
plt.xlabel("GNI per Capita ($)")
plt.ylabel("HDI")
plt.title("Outlier Detection: GNI per Capita vs HDI (South Asia, 2022)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#7. Outlier Detection:
#• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using the 1.5 × IQR rule.
#Detecting outliers for South Asia:
southasian_outliers = hdi_southasia_2020_2022.copy()

Q1_southasian_hdi = southasian_outliers["hdi"].quantile(0.25)
Q3_southasian_hdi = southasian_outliers["hdi"].quantile(0.75)
IQR_southasian_hdi = Q3_southasian_hdi - Q1_southasian_hdi

southasian_hdi_lower = Q1_southasian_hdi - 1.5 * IQR_southasian_hdi
southasian_hdi_upper = Q3_southasian_hdi + 1.5 * IQR_southasian_hdi

southasian_outliers["is_hdi_outlier"] = (
    (southasian_outliers["hdi"] < southasian_hdi_lower) |
    (southasian_outliers["hdi"] > southasian_hdi_upper)
)

print("South Asia - HDI outliers:")
print(southasian_outliers[southasian_outliers["is_hdi_outlier"]][
    ["country", "year", "hdi"]
])

# ---- GNI per Capita (South Asia) ----
Q1_southasian_gni = southasian_outliers["gross_inc_percap"].quantile(0.25)
Q3_southasian_gni = southasian_outliers["gross_inc_percap"].quantile(0.75)
IQR_southasian_gni = Q3_southasian_gni - Q1_southasian_gni

southasian_gni_lower = Q1_southasian_gni - 1.5 * IQR_southasian_gni
southasian_gni_upper = Q3_southasian_gni + 1.5 * IQR_southasian_gni

southasian_outliers["is_gni_outlier"] = (
    (southasian_outliers["gross_inc_percap"] < southasian_gni_lower) |
    (southasian_outliers["gross_inc_percap"] > southasian_gni_upper)
)

print("\nSouth Asia - GNI outliers:")
print(southasian_outliers[southasian_outliers["is_gni_outlier"]][
    ["country", "year", "gross_inc_percap"]
])
```

```python
#Detecting outliers for Middle East:
middleeast_outliers = hdi_middleeast_2020_2022.copy()

# ---- HDI (Middle East) ----
Q1_middleeast_hdi = middleeast_outliers["hdi"].quantile(0.25)
Q3_middleeast_hdi = middleeast_outliers["hdi"].quantile(0.75)
IQR_middleeast_hdi = Q3_middleeast_hdi - Q1_middleeast_hdi

middleeast_hdi_lower = Q1_middleeast_hdi - 1.5 * IQR_middleeast_hdi
middleeast_hdi_upper = Q3_middleeast_hdi + 1.5 * IQR_middleeast_hdi

middleeast_outliers["is_hdi_outlier"] = (
    (middleeast_outliers["hdi"] < middleeast_hdi_lower) |
    (middleeast_outliers["hdi"] > middleeast_hdi_upper)
)

print("\nMiddle East - HDI outliers:")
print(middleeast_outliers[middleeast_outliers["is_hdi_outlier"]][
    ["country", "year", "hdi"]
])

# ---- GNI per Capita (Middle East) ----
Q1_middleeast_gni = middleeast_outliers["gross_inc_percap"].quantile(0.25)
Q3_middleeast_gni = middleeast_outliers["gross_inc_percap"].quantile(0.75)
IQR_middleeast_gni = Q3_middleeast_gni - Q1_middleeast_gni

middleeast_gni_lower = Q1_middleeast_gni - 1.5 * IQR_middleeast_gni
middleeast_gni_upper = Q3_middleeast_gni + 1.5 * IQR_middleeast_gni

middleeast_outliers["is_gni_outlier"] = (
    (middleeast_outliers["gross_inc_percap"] < middleeast_gni_lower) |
    (middleeast_outliers["gross_inc_percap"] > middleeast_gni_upper)
)

print("\nMiddle East - GNI outliers:")
print(middleeast_outliers[middleeast_outliers["is_gni_outlier"]][
    ["country", "year", "gross_inc_percap"]
])
```
